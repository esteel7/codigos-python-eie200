# Métodos Numéricos para la Solución de Ecuaciones Diferenciales

En este notebook, se explicarán tres métodos numéricos de solución de ecuaciones diferenciales ordinarias:
1. **Método de Euler**
2. **Método de Heun** (o Euler mejorado)
3. **Método del Punto Medio**

Todos los métodos se aplicarán para resolver la siguiente ecuación diferencial:

$$
\frac{dy}{dx} = -2x^3 + 12x^2 - 20x + 8.5
$$

Además, se compararán los resultados numéricos con la solución exacta:

$$
y(x) = -0.5x^4 + 4x^3 - 10x^2 + 8.5x + 1.0
$$

## Método de Euler

El método de Euler es uno de los métodos numéricos más simples para resolver ecuaciones diferenciales ordinarias de la forma:

$$
\frac{dy}{dx} = f(x, y)
$$

con una condición inicial $$y(x = 0) = y_0$$. El método de Euler aproxima la solución al avanzar paso a paso en el intervalo de integración, utilizando una pendiente calculada en el punto inicial de cada paso.

### Fórmula del Método de Euler

Dado un paso $h$ (el tamaño del intervalo entre dos puntos consecutivos), el valor de $y$ en el siguiente paso, $y_{i+1}$, se calcula a partir del valor actual $y_i$ mediante la fórmula:

$$
y_{i+1} = y_i +  f(x_i, y_i) \cdot h

$$


Donde:
- $x_i$ es el valor actual de $x$,
- $y_i$ es el valor actual de $y$,
- $f(x_i, y_i)$ es la derivada o pendiente en el punto $(x_i, y_i)$,
- $h$ es el tamaño del paso, vale decir $x_{i+1} - x_i$.

Este proceso se repite desde $x_0$ hasta el valor final de $x$, avanzando en intervalos de tamaño $h$.

### Ventajas y Desventajas

- **Ventajas**: Es un método simple y fácil de implementar.
- **Desventajas**: Puede ser ineficiente y poco preciso para problemas donde se requiere mayor exactitud, o cuando se utilizan pasos de integración grandes, ya que es un método de primer orden (el error global es proporcional a $h$.

### Ejemplo


```python
import numpy as np
import matplotlib.pyplot as plt

# Función que representa la ecuación diferencial dy/dx = f(x, y)
def dydx(x,y):
    return -2*x**3 + 12*x**2 - 20*x + 8.5

# Solución exacta
def y_exacta(x):
    return -0.5*x**4 + 4*x**3 - 10*x**2 + 8.5*x + 1.0

# Definimos el paso
h = 0.1

# Condición inicial y(x = 0) = 1.0
y0 = 1.0

# Inicializamos los arreglos x e y_euler
x_range = [0.0, 4.0]
x = np.arange(x_range[0], x_range[1] + h, h)
y_euler = np.zeros_like(x)
y_euler[0] = y0

# Resolvemos la EDO usando el método de Euler
for i in range(x.shape[0] - 1):
    k1 = dydx(x[i], y_euler[i])
    y_euler[i+1] = y_euler[i] + k1 * h

# Definimos la solución de referencia
y_ref = y_exacta(x)

# Graficamos las soluciones
plt.plot(x, y_euler)
plt.plot(x, y_ref)
plt.legend(["Euler", "Referencia"])
plt.grid(True)
plt.show()
```
